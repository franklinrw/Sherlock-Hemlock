```python
models = ["de_core_news_sm", "en_core_web_sm", "es_core_news_sm", "fr_core_news_sm", "it_core_news_sm", "nl_core_news_sm", "pt_core_news_sm"]
```

```python
dataset = load_dataset("joelito/mapa")
```

```python
df_train = pd.DataFrame(dataset["train"].to_pandas())
df_val = pd.DataFrame(dataset["validation"].to_pandas())
df_test = pd.DataFrame(dataset["test"].to_pandas())
```

```python
# List of language codes to drop
languages_to_keep = ["de", "en", "es", "fr", "it", "nl", "pt"]
```

```python
# Drop the rows with the specified language codes
df_train = df_train[df_train['language'].isin(languages_to_keep)]
df_val = df_val[df_val['language'].isin(languages_to_keep)]
df_test = df_test[df_test['language'].isin(languages_to_keep)]

# Reset the index of the filtered DataFrame
df_train.reset_index(drop=True, inplace=True)
df_val.reset_index(drop=True, inplace=True)
df_test.reset_index(drop=True, inplace=True)

# Drop the columns not required for further processing
df_train = df_train.drop(["type", "file_name", "sentence_number"], axis=1)
df_val = df_val.drop(["type", "file_name", "sentence_number"], axis=1)
df_test = df_test.drop(["type", "file_name", "sentence_number"], axis=1)
```

```python
def extract_pos_dep_head(sentence, nlp):
    try:
        doc = nlp(" ".join(sentence))
        pos = [token.pos_ for token in doc]
        dep = [token.dep_ for token in doc]
        heads = [token.head.i for token in doc]
    except:
        pos, dep, heads = [], [], []
    return pos, dep, heads
```

```python
def process_languages(df, models):
    df['pos'] = ''
    df['dep'] = ''
    df['heads'] = ''
    for model in models:
        lang = model.split("_")[0]
        nlp = spacy.load(model)
        lang_df = df.loc[df['language'] == lang].copy()
        lang_df[['pos', 'dep', 'heads']] = lang_df['tokens'].apply(lambda x: pd.Series(extract_pos_dep_head(x, nlp)))
        df.loc[df['language'] == lang, ['pos', 'dep', 'heads']] = lang_df[['pos', 'dep', 'heads']].values
    return df
```

```python
df_train = process_languages(df_train, models)
df_val = process_languages(df_val, models)
df_test = process_languages(df_test, models)
```
